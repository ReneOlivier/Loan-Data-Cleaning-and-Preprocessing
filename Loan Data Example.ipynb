```python
# importing numpy package

import numpy as np
np.set_printoptions(suppress = True, linewidth = 100, precision = 2)
```

```python
#importing the csv file that I'll be working with - with absolute file directory - makes sense to me to use that method.
#Initiailly attempted a np.loadtxt, errors occured, so we are using np.genfromtxt

raw_data_np = np.genfromtxt("Loan Data\\loan-data.csv",
                            delimiter = ";",
                            skip_header = 1,
                            autostrip = True)
raw_data_np
```

### **Checking for Incomplete Data**

```python
# Let's see how much data is missing from the dataset

np.isnan(raw_data_np).sum()
```

```python
# Creating 2 temporary values 

temporary_fill = np.nanmax(raw_data_np) + 1 # Stores rounded result and will serve as a filler for all missing values
temporary_mean = np.nanmean(raw_data_np, axis = 0) # Stores the mean for every column
```

```python
# Checking how many columns contain empty/text values

temporary_mean
```

```python
# Numpy does not like storing multiple types of data, so we should split it into smaller arrays

temporary_stats = np.array([np.nanmin(raw_data_np, axis = 0),
                           temporary_mean,
                           np.nanmax(raw_data_np, axis = 0)])
```

```python
# Displaying the new variable, we can now see min,mean and max values

temporary_stats 
```

### **Splitting the Dataset**

#### Splitting the columns

```python
# Before we split dataset, we first need to know which columns contain strings and which contain numeric values
# This will return value to show indexes of columns that only contain strings
column_string = np.argwhere(np.isnan(temporary_mean)).squeeze()
column_string
```

```python
# This will return value to show indexes of columns that only contain numeric values

column_numeric = np.argwhere(np.isnan(temporary_mean) == False).squeeze()
column_numeric
```

#### Re-importing the Dataset

```python
# We copy our code for importing, just change the variable name and add the usecols argument, setting the dtype to string

loan_data_strings = np.genfromtxt("Loan Data\\loan-data.csv",
                            delimiter = ";",
                            skip_header = 1,
                            autostrip = True,
                            usecols = column_string,
                            dtype = str)
loan_data_strings
```

```python
# Copy code to import numeric values, add the usecols argument and filling_values argument

loan_data_numeric = np.genfromtxt("Loan Data\\loan-data.csv",
                            delimiter = ";",
                            skip_header = 1,
                            autostrip = True,
                            usecols = column_numeric,
                            filling_values = temporary_fill) # Temp fill added to know which values are missing
loan_data_numeric
```

### The names of the columns

```python
#We need to store the header of each column before we can start working with the data, otherwise we lose track of the data we are storing

header_full = np.genfromtxt("Loan Data\\loan-data.csv",
                            delimiter = ";",
                            skip_footer = raw_data_np.shape[0],# This line tells the function to remove all rows after the header
                            autostrip = True,
                            dtype = str)
header_full
```

```python
# Now lets split the headers into 2 variables each one holding the names of the columns for both strings and numeric

header_strings, header_numeric = header_full[column_string], header_full[column_numeric]
```

```python
# Test to see that the columns are correct

header_strings, header_numeric
```

### **Creating Checkpoints:**

```python
# create checkpoint to store a copy of our code to avoid losing progress
# also useful if preprocessing many parts of a dataset - a failsafe we can rely on

def checkpoint(file_name, checkpoint_header, checkpoint_data):
    np.savez(file_name, header = checkpoint_header, data = checkpoint_data)
    checkpoint_variable = np.load(file_name + ".npz")
    return (checkpoint_variable)
```

```python
checkpoint_test = checkpoint("checkpoint-test", header_strings,loan_data_strings)
```

### **Issue Date**

```python
np.unique(loan_data_strings[:, 0 ]) # output consists of an empty space that represents missing data
# as well as 12 other strings for the months arranged alphabetically.
# regarding the year - all the loans seem to be from the same year and can safely be stripped without losing insightful information
```

```python
loan_data_strings[:, 0] = np.chararray.strip(loan_data_strings[:, 0], "-15") # code does not overide the data unless setting the variable
```

```python
np.unique(loan_data_strings[:, 0 ])
```

```python
# Common practice displays months as numeric values - to do this we first create a list or an array of the months in order

months = np.array(['', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
# Then we create a loop to cycle through and associate a numeric equivalent with each one and missing values to zero
# to seperate it from the rest because we want to attribute the annual values to candidates who have no issue dats on record

for i in range(13):
    loan_data_strings[:, 0] = np.where(loan_data_strings[:, 0] == months[i],
                                       str(i),
                                       loan_data_strings[:, 0])
```
